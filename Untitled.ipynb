#### 分析1:浮点型数据70列，整数型数据13列，object 型7列，其中student_feature缺失值多

### 2.无关特征删除

```python
data = data.drop(['Unnamed: 0','custid','trade_no','bank_card_no','student_feature','source','id_name','latest_query_time','loans_latest_time'],axis=1)
data.info()
```

### 3.数据类型处理
#### 由于删除后，reg_preference_for_trad 为object类型数据，需处理，本次采用独热编码


```python
data = pd.get_dummies(data,columns=["reg_preference_for_trad"])
data = data.convert_objects(convert_numeric=True)
```

```python
data.info()
```

### 4.缺失值处理
####  对于缺失值，采取了均值的简单方法

```python
data = data.fillna(data.mean())
```
